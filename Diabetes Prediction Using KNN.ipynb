```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes.csv')
print(len(df))
df
```

```python
df.columns=['Pregnancies Week','Glucose Percentage','Blood Pressure',
            'Skin Thickness','Insulin Percentage','BMI','Diabetes Pedigree Function','Age','Outcome']
```

```python
df.head()
```

```python
#check for nullity
df.isnull().sum()
```

# Data Understanding
we can't have zeros in (Glucose Percentage ,	Blood Pressure 	,Skin Thickness 	,Insulin Percentage ,	BMI )


```python
# Replace zeros
zero_not_acepted = ['Glucose Percentage','Blood Pressure',
            'Skin Thickness','Insulin Percentage','BMI']
```

```python
for col in zero_not_acepted:
    df[col] = df[col].replace(0,np.NAN)
    mean = df[col].mean(skipna = True)
    df[col] = df[col].replace(np.NAN,mean)
    
```

```python
df
```

```python
#split the data into features and labels
X = df.iloc[:,0:8]
y = df.iloc[:,8]
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#normalization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import math

help(KNeighborsClassifier)
math.sqrt(len(y_test))
```

```python
#knn model
knn = KNeighborsClassifier(n_neighbors=13,p=2, metric='minkowski')
knn.fit(X_train,y_train)
```

```python
# predict the test set results
y_pred = knn.predict(X_test)
y_pred
```

```python
#Evalution
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
